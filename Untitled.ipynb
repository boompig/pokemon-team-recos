```python
import json
fname = "data/teams/all-teams.json"
f = open(fname)
teams = json.load(f)
f.close()
teams
```

```python
# what's the most common pokemon on a team?

def get_pokemon_frequencies(teams):
    freq = {}
    for team in teams:
        for mon in team:
            freq.setdefault(mon, 0)
            freq[mon] += 1
    return freq

freq = get_pokemon_frequencies(teams)
freq
```

```python
for mon, count in sorted(freq.items(), key=lambda item: item[1], reverse=True):
    print("{} -> {}".format(mon, count))
```

```python
import matplotlib.pyplot as plt

def graph_frequencies(freq, title):
    plt.rcParams['figure.dpi'] = 200
    plt.rcParams['figure.figsize'] = [20, 30]

    mons = []
    counts = []
    # for mon, count in sorted(freq.items(), key=lambda item: item[1], reverse=True):
    for mon, count in sorted(freq.items(), key=lambda item: item[1]):
        mons.append(mon)
        counts.append(count)
        
    plt.title(title, fontsize=18)

    plt.barh(y=mons, width=counts)

    # plt.xticks(mons, rotation=90)
    plt.yticks(mons, fontsize=14)

    plt.autoscale(enable=True, axis='y', tight=True)

    plt.show()
```

```python
%matplotlib inline

graph_frequencies(freq, "Frequency of Pokemon among all teams")
```

```python
# different teams for different starters

starters = ["sceptile", "swampert", "blaziken"]

starter_teams = {}

for team in teams:
    if team == []:
        continue
    no_starter = True
    for starter in starters:
        if starter in team:
            starter_teams.setdefault(starter, [])
            starter_teams[starter].append(team)
            no_starter = False
            
    if no_starter:
        print("no starter for team {}".format(team))
```

```python
freq = get_pokemon_frequencies(starter_teams["swampert"])
freq
```

```python
# remove swampert
freq.pop("swampert")

graph_frequencies(freq, "frequency of Pokemon among Swampert teams")
```

```python
freq = get_pokemon_frequencies(starter_teams["blaziken"])
freq.pop("blaziken")
graph_frequencies(freq, "frequency of Pokemon among Blaziken teams")
```

```python
# which pairs of pokemon occur most frequently
def get_all_pokemon(teams):
    mons = set([])
    for team in teams:
        for mon in team:
            mons.add(mon)
    return mons

def get_freq_pairs(teams):
    pairs = {}
    for team in teams:
        for mon1 in team:
            for mon2 in team:
                if mon2 <= mon1:
                    continue
                k = (mon1, mon2)
                pairs.setdefault(k, 0)
                pairs[k] += 1
    return pairs

all_mons = get_all_pokemon(teams)
pairs = get_freq_pairs(teams)

pairs
```

```python
plt.rcParams['figure.dpi'] = 200
plt.rcParams['figure.figsize'] = [20, 50]

mons = []
counts = []
# for mon, count in sorted(freq.items(), key=lambda item: item[1], reverse=True):
for mon, count in sorted(pairs.items(), key=lambda item: item[1]):
    if count > 2:
        mons.append(str(mon))
        counts.append(count)
    
#print(mons)

plt.title("frequency of pairs", fontsize=18)

plt.barh(y=[str(mon) for mon in mons], width=counts)

# plt.xticks(mons, rotation=90)
plt.yticks(mons, fontsize=14)

plt.autoscale(enable=True, axis='y', tight=True)

plt.show()
```

```python
import networkx as nx

plt.rcParams['figure.dpi'] = 200
plt.rcParams['figure.figsize'] = [20, 30]

G = nx.Graph()

for mon in all_mons:
    G.add_node(mon)
    
for (mon1, mon2), count in pairs.items():
    G.add_edge(mon1, mon2, weight=count)

# get the number of connected components
print(nx.number_connected_components(G))
largest_cc = max(nx.connected_components(G), key=len)
subgraph = G.subgraph(largest_cc)
    
#pos = nx.spring_layout(G, scale=1, weight='weight')
pos = nx.spring_layout(subgraph, scale=1, weight='weight')

#nx.draw_networkx_nodes(G, pos, node_size=700)
#nx.draw_networkx_edges(G, pos, edgelist=elarge,
#                       width=6)
#nx.draw_networkx_labels(G, pos, font_size=20, font_family='sans-serif')




#nx.draw_networkx(G, pos)
nx.draw_networkx(subgraph, pos, edge_color='#444444')

plt.axis('off')


# save this to file
plt.savefig("figures/teams-graph.png", dpi=300)

plt.show()
```
